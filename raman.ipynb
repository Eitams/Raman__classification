# Deep Learning Pipeline

Assuming our dataset, task, and model are defined, the pipeline is basically always as follows:
1. Create the model
2. Load data and prepare the samples for the model
3. Initialize the optimizer and training hyperparameters

And then the training loop
4. Get a batch of data from the dataset
5. Move data to GPU
6. Do the forward pass to compute the output
7. Compute the loss
8. Compute the gradients with backpropagation (i.e., the forward pass)
9. Update the model parameters 
10. Go to 4) if the model has not converged (or other stopping criterion)

```python
def findConv1dOutShape(win, conv):
    # Get conv arguments
    kernel_size = conv.kernel_size[0]
    stride = conv.stride[0]
    padding = conv.padding[0]
    dilation = conv.dilation[0]

    wout = np.floor((win - kernel_size + 2 * padding) / stride + 1)

    return int(wout)

def findpollingOutShape(win, conv):
    # Get conv arguments
    kernel_size = conv.kernel_size
    stride = conv.stride
    padding = conv.padding
    dilation = conv.dilation

    wout = np.floor((win - kernel_size ) //stride + 1)

    return int(wout)
```

```python
def get_metrics(preds, golds, num_class, args):
    metrics = {}

    metrics['accuracy'] = sklearn.metrics.accuracy_score(y_true=golds, y_pred=preds)
    metrics['confusion_matrix'] = sklearn.metrics.confusion_matrix(y_true=golds, y_pred=preds)
    avg_type = 'binary'
    metrics['precision_{}'.format(avg_type)] = sklearn.metrics.precision_score(y_true=golds, y_pred=preds, average=avg_type)
    metrics['recall_{}'.format(avg_type)] = sklearn.metrics.recall_score(y_true=golds, y_pred=preds, average=avg_type)
    metrics['f1_{}'.format(avg_type)] = sklearn.metrics.f1_score(y_true=golds, y_pred=preds, average=avg_type)

    # Clean the binary
    if num_class == 2:
        for k in list(metrics.keys()):
            if k.endswith('_binary'):
                metrics[k.replace('_binary', '')] = metrics.pop(k)

    return metrics
```

```python
import torch.optim as optim
import copy
from torch.optim.lr_scheduler import ReduceLROnPlateau

# Set the device to GPU if available, otherwise CPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
# Initialize the model
model = RamanModel(single_input.shape[2], args).to(device)

# Define the loss function and optimizer
criterion = nn.BCEWithLogitsLoss()
#criterion = nn.BCELoss()
#optimizer = optim.Adam(model.parameters(), lr=args.init_lr)
optimizer = optim.Adam(model.parameters(), lr=0.1)
lr_scheduler = ReduceLROnPlateau(optimizer, mode='min',factor=0.5, patience=20,verbose=1)



best_epoch = 0
best_val_so_far = float('inf')
test_perf = 0
best_model_wts = copy.deepcopy(model.state_dict())

## initialize lists of loss per epoch to use in model loss plot
train_loss_perEpoch = []
train_accuracy_perEpoch = []
val_loss_perEpoch = []
val_accuracy_perEpoch = []

# Training loop
for epoch in range(args.epochs):
    print('------------------------------------------------------------------------')
    print('Epoch: {}'.format(epoch))
    current_lr = optimizer.param_groups[0]['lr']

    ##########
    ## Train
    ##########
   
    model.train()
    train_loss_perBatch = []
    train_accuracy_perBatch = []

    running_loss = 0.0

    for batch in tqdm.tqdm(train_dataloader, desc='Training'):
    # for inputs, labels in train_dataloader:  # dataloader represents your training data
        # move them to GPU
        inputs = batch['features'].to(device)
        target = batch['target'].to(device) #BxL      
       
        #######################
        ## Forward pass
        #######################

        outputs = model(inputs)
        loss = criterion(outputs.float(), target.float())   
        # Add L1 and L2 regularization to the loss function
        l1_lambda = 0.01  # Adjust this value for L1 regularization strength
        l2_lambda = 0.01  # Adjust this value for L2 regularization strength
        l1_reg = torch.tensor(0.)
        l2_reg = torch.tensor(0.)
        # print(loss)
        for param in model.parameters():
            l1_reg += torch.norm(param, 1)
            l2_reg += torch.norm(param, 2)

        loss = loss + l1_lambda * l1_reg + l2_lambda * l2_reg 
        train_loss_perBatch.append(loss.item())            
        running_loss += loss.item()
        metrics = get_metrics(outputs, target, 2, args)
        train_accuracy_perBatch.append(metrics["accuracy"])

        #######################
        ## Backward pass
        #######################        
        optimizer.zero_grad()
        loss.backward()
        # Update wights
        optimizer.step()
        
        
    train_loss_perEpoch.append(running_loss / float(len(train_dataloader.dataset))) # divide by number of samples
    #train_loss_perEpoch.append(running_loss / (len(train_loss_perBatch) * args.batch_size)) # divide by number of samples
    train_accuracy_perEpoch.append(np.mean(metrics["accuracy"]))

    ##############
    ## Validation
    ##############

    val_loss_perBatch = []
    test_f1_perBatch = []
    test_recall_perBatch = []
    test_conf_perBatch = []
    test_precision_perBatch = []
    val_accuracy_perBatch = []
    model.eval()
    
    # We do not store information relative to gradients as we do not update the model.
    with torch.no_grad():
        running_loss = 0.0
        run_acc = 0.0
        for batch in tqdm.tqdm(test_dataloader, desc='Validation'):
            
            # move them to GPU
            inputs = batch['features'].to(device)
            target = batch['target'].to(device) #BxL  

           ##################
           ## Forward pass
           ##################

            outputs = model(inputs)
            loss = criterion(outputs.float(), target.float())   
            # Add L1 and L2 regularization to the loss function
            l1_lambda = 0.01  # Adjust this value for L1 regularization strength
            l2_lambda = 0.01  # Adjust this value for L2 regularization strength
            l1_reg = torch.tensor(0.)
            l2_reg = torch.tensor(0.)
            # print(loss)
            for param in model.parameters():
                l1_reg += torch.norm(param, 1)
                l2_reg += torch.norm(param, 2)

            loss = loss + l1_lambda * l1_reg + l2_lambda * l2_reg       
            running_loss += loss.item()
            val_loss_perBatch.append(loss.item())


            metrics = get_metrics(outputs, target, 2, args)
            #test_f1_perBatch.append(metrics["f1"])
            #test_precision_perBatch.append(metrics["precision"])
            val_accuracy_perBatch.append(metrics["accuracy"])
            #test_recall_perBatch.append(metrics["recall"])
            test_conf_perBatch.append(metrics["confusion_matrix"])
        
        val_loss_perEpoch.append(running_loss / float((len(test_dataloader.dataset)))) # divide by number of validation samples
        #val_accuracy_perEpoch.append(run_acc / float((len(test_dataloader.dataset)))) # divide by number of validation samples
        #val_accuracy_perEpoch.append(np.mean(metrics["accuracy"])) # divide by number of validation samples
        val_accuracy_perEpoch.append(np.mean(val_accuracy_perBatch)) # divide by number of validation samples




        
        # print(f"Epoch {epoch}/{args.epochs}, Loss: {epoch_loss:.4f}, Validation Fbeta: {fbeta(outputs, labels):.4f}")
    
    # Monitoring
    print()
    print('Train -  loss: {:.4f} Accuracy: {:.4f}'.format(train_loss_perEpoch[epoch], train_accuracy_perEpoch[epoch]))
    print('Validation - loss: {:.4f} Accuracy: {:.4f}'.format(val_loss_perEpoch[epoch], val_accuracy_perEpoch[epoch]))
    print()
    
    print("Confusion materix:")
    print(sum(test_conf_perBatch))
    print()
    #test_precision = np.mean(test_precision)
    #print('Test precision: {:.4f}'.format(test_precision))
    #test_accuracy = np.mean(test_accuracy)
    #print('Test accuracy: {:.4f}'.format(test_accuracy))
    #test_f1 = np.mean(test_f1)
    #print('Test  Macro F1: {:.4f}'.format(test_f1))
    
    #test_recall = np.mean(test_recall)
    #print('Test  recall: {:.4f}'.format(test_recall))

    ## Store best model
    if best_val_so_far > val_loss_perEpoch[epoch]:
        best_val_so_far = val_loss_perEpoch[epoch]
        best_epoch = epoch
        best_model_wts = copy.deepcopy(model.state_dict())
        
        # store weights into a local file
        torch.save(model.state_dict(), "Models_weights/weights.pt")
        if(args.verbose):
            print("Copied best model weights!")
        
    # learning rate schedule
    
    lr_scheduler.step(val_loss_perEpoch[epoch])
    if current_lr != optimizer.param_groups[0]['lr']:
        if(args.verbose):
            print("Loading best model weights!")
        model.load_state_dict(best_model_wts)
    


    print('Best Epoch: {}, Validation loss {:.4f}, Validation accuracy: {:.4f}'.format(best_epoch, best_val_so_far, val_accuracy_perEpoch[best_epoch]))
    print()
    print()
    # # Calculate average loss for the epoch
    # epoch_loss = running_loss / len(train_dataloader)
    # print(f"Epoch {epoch+1}/{args.epochs}, Loss: {epoch_loss:.4f}")

# load best model weights
#model.load_state_dict(best_model_wts)
```
